```python
# folium은 Leaflet.js를 이용해서 위치 좌표계 값을 지도위에 시각화 할 수 있다.
# !pip install folium
import folium
```

```python
# folium 라이브러리의 Map() 함수의 location 속성으로 지정한 위치를 중심으로 하는 zoom_start 옵션으로 지정한 배율을 사용하는
# 지도를 만든다.
# folium.Map(location=[위도, 경도], zoom_start=배율, tiles='지도 스타일')
map_folium = folium.Map(location=[37.50077723677026, 127.0372256980366], zoom_start=17)
# folium의 save() 함수로 작성된 지도를 html 파일로 저장할 수 있다.
map_folium.save('./output/map.html')
map_folium
```

```python
# 지도 스타일: Open Street Map(기본값), Stamen Terrain, Stamen Toner
map_folium = folium.Map(location=[37.50077723677026, 127.0372256980366], zoom_start=17, tiles='Open Street Map')
map_folium
```

```python
map_folium = folium.Map(location=[37.56998576305384, 126.98366305385956], zoom_start=16)

# folium 라이브러리의 Marker() 함수로 location 속성으로 지정한 위치에 마커를 만든다.
# popup 속성으로 마커를 클릭했을 때 표시할 문자열을 지정할 수 있다.
# 작성된 마커는 add_to() 함수로 지도에 추가한다.
folium.Marker(location=[37.569977259279824, 126.9837059692024], popup='보신각').add_to(map_folium)
folium.Marker(location=[37.56648227185431, 126.97791291153096], popup='서울특별시청').add_to(map_folium)
folium.Marker(location=[37.5658724427688, 126.97513433748387], popup='덕수궁').add_to(map_folium)
folium.Marker(location=[37.57981254103927, 126.97701954036695], popup='경복궁').add_to(map_folium)
folium.Marker(location=[37.57605939756449, 126.97689493338027], popup='광화문').add_to(map_folium)
folium.Marker(location=[37.56995695553171, 126.98228391153103], popup='영풍문고 - 종로본점').add_to(map_folium)

map_folium.save('./output/map.html')
map_folium
```

```python
map_folium = folium.Map(location=[37.56998576305384, 126.98366305385956], zoom_start=16)

# 마커를 클릭했을 때 팝업되는 말풍선의 크기가 마음에 들지 않으면 folium의 Popup() 함수의 max_width 속성으로 변경하면 된다.
popup = folium.Popup('보오오오오오오오시인ㄴㄴㄴㄴㄴㄴㄴㄴㄴ각ㄱㄱㄱㄱㄱㄱㄱㄱㄱㄱㄱㄱㄱㄱㄱ', max_width=200)
folium.Marker(location=[37.569977259279824, 126.9837059692024], popup=popup).add_to(map_folium)

map_folium
```

```python
map_folium = folium.Map(location=[37.56998576305384, 126.98366305385956], zoom_start=16)

# folium의 icon 속성에 Icon() 함수를 사용해서 마커의 색상과 아이콘 모양을 변경할 수 있다.
# Icon() 함수의 color 속성으로 지정할 수 있는 색상은 아래와 같다.
# white', 'lightred', 'green', 'pink', 'orange', 'beige', 'darkred', 'darkgreen', 'cadetblue', 'darkpurple', 'lightgray', 
#'purple', 'black', 'blue', 'red', 'lightblue', 'gray', 'darkblue', 'lightgreen'
# 마커에 표시되는 아이콘은 https://getbootstrap.com/docs/3.3/components/#glyphicons-glyph에서 제공하는 아이콘을 사용한다.

folium.Marker(location=[37.569977259279824, 126.9837059692024], popup=folium.Popup('보신각', max_width=100),
             icon=folium.Icon(color='pink', icon='star')).add_to(map_folium)
folium.Marker(location=[37.56648227185431, 126.97791291153096], popup=folium.Popup('서울특별시청', max_width=100),
             icon=folium.Icon(color='purple', icon='star-empty')).add_to(map_folium)
folium.Marker(location=[37.5658724427688, 126.97513433748387], popup=folium.Popup('덕수궁', max_width=100),
             icon=folium.Icon(color='cadetblue', icon='tower')).add_to(map_folium)

map_folium
```

```python
map_folium = folium.Map(location=[37.56998576305384, 126.98366305385956], zoom_start=16)

# folium의 CircleMarker() 함수로 원 모양의 마커를 만들 수 있다.
# radius 속성으로 원의 크기를 지정할 수 있고 color 속성으로 원 색상을 fill_color 속성으로 원 내부 색상을 지정할 수 있다.
folium.CircleMarker(location=[37.569977259279824, 126.9837059692024], popup=folium.Popup('보신각', max_width=100),
             radius=100, color='#B93CE8', fill_color='#0DFB78').add_to(map_folium)

map_folium
```
