```python
data = [9, 1, 3, 4, 8]
for i in range(len(data) - 1):
    #  => ()    flag  0 .
    flag = 0
    for j in range(len(data) - 1 - i):
        if data[j] > data[j + 1]:
            data[j], data[j + 1] = data[j + 1], data[j] #   => 
            #  =>     flag   1 .
            flag = 1
    # ===== for j =>  
    #  =>               
    #    . =>  .
    if flag == 0:
        print('{}   {}    '.format(len(data) - 1, i + 1))
        break
    print('{}  : {}'.format(i + 1, data))
# ===== for i =>  
# print(' : {}'.format(data))
```
