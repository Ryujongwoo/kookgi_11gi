```python
data = [9, 1, 3, 4, 8]
for i in range(len(data) - 1):
    # ⓐ => 회전(정렬) 작업이 시작되기 전에 flag 변수를 0으로 초기화시킨다.
    flag = 0
    for j in range(len(data) - 1 - i):
        if data[j] > data[j + 1]:
            data[j], data[j + 1] = data[j + 1], data[j] # 데이터 교환 => 정렬중
            # ⓑ => 데이터 교환 작업이 실행되면 flag 변수의 값을 1로 수정한다.
            flag = 1
    # ===== for j => 회전 종료
    # ⓒ => 한 회전이 종료되었을 때 데이터 교환 작업이 한번도 실행되지 않았다면 현재 데이터는 모두 정렬이 되어있는
    # 상태를 의미하므로 정렬을 중지시킨다. => 반복을 탈출한다.
    if flag == 0:
        print('{} 회전을 해야하지만 {} 회전 후 중간 종료됨'.format(len(data) - 1, i + 1))
        break
    print('{} 회전 결과: {}'.format(i + 1, data))
# ===== for i => 정렬 종료
# print('정렬 결과: {}'.format(data))
```
