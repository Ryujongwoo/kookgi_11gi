삽입 정렬은 이미 정렬된 파일에 새로운 데이터(레코드)를 위치에 맞게 삽입한다.  
정렬할 데이터가 n개일 경우 회전수는 n-1번이 된다. => 데이터가 5개면 회전수는 4번이다.

```python
from IPython.display import Image
Image('./insertionSort.png', width='1000')
```

```python
for i in range(1, 5): # 회전수, 정렬된 파일에 삽입할 데이터의 위치
    for j in range(i - 1, -1, -1): # 삽입할 데이터와 비고하기 위해 정렬된 파일에서 데이터를 선택하는 위치
        print('[i = {}, j = {}]'.format(i, j), end=' ')
    print()
```

```python
data = [8, 3, 4, 9, 1]
for i in range(1, len(data)):
    # 이미 정렬된 파일에 새 데이터가 삽입될 위치를 찾기 위해 새 데이터와 인덱스를 각각 key와 index에 저장한다.
    key = data[i]
    index = i
    # 새 데이터가 삽입될 위치를 찾는다.
    for j in range(i - 1, -1, -1):
        if data[j] > key:
            # 이미 정렬된 파일의 데이터가 크기때문에 정렬된 파일의 데이터를 인덱스가 1큰 위치로 이동시키고 삽입할
            # 데이터의 인덱스 값을 1감소 시킨다.
            data[j + 1] = data[j]
            index -= 1
        else:
            # 삽입할 데이터가 이미 정렬된 파일의 데이터와 같거나 더 크다는 의미이므로 이미 정렬된 데이터 앞에 
            # 위치한 데이터와 더 이상 비교할 필요가 없으므로 반복해서 비교할 필요가 없어진다. => 반복을 탈출한다.
            break
        # ===== if data[j] > key
    # ===== for j => 회전 종료
    # 새 데이터를 정렬된 파일의 위치에 맞게 삽입한다.
    data[index] = key
    print('{} 회전 결과: {}'.format(i, data))
# ===== for i => 정렬 종료`
# print('정렬 종료: {}'.format(data))
```
