```python
row, col = map(int, input('2차원 리스트를 만들려하는 행과 열의 개수: ').split())
data = [[0] * col for i in range(row)]

n = 0 # 리스트에 1씩 증가하며 채워질 숫자가 기억되는 변수를 선언한다.
# 전체 행의 개수만큼 반복한다.
for i in range(len(data)):
    # 짝수행은 숫자가 왼쪽에서 오른쪽으로 채워지고 홀수행은 숫자가 오른쪽에서 왼쪽으로 채워진다.
    # 짝수행인가 홀수행인가 판단한다.
    if i % 2 == 0:
        # 짝수행은 왼쪽에서 오른쪽으로 숫자를 채운다. j => 0, 1, 2, 3, 4
        for j in range(len(data[0])):
            n += 1
            data[i][j] = n
    else:
        # 홀수행은 오른쪽에서 왼쪽으로 숫자를 채운다. j => 4, 3, 2, 1, 0
        for j in range(len(data[0]) - 1, -1, -1):
            n += 1
            data[i][j] = n

for i in range(row):
    for j in range(col):
        print('{0:3d} '.format(data[i][j]), end='')
    print()
```

```python
row, col = map(int, input('2차원 리스트를 만들려하는 행과 열의 개수: ').split())
data = [[0] * col for i in range(row)]

n = 0
start = 0     # 2차원 리스트에 숫자가 채워지기 시작하는 열의 인덱스
end = col - 1 # 2차원 리스트에 숫자가 채워지는 마지막 열의 인덱스
sw = 1        # start 부터 end 까지 증가치로 사용할 변수

for i in range(row):
    for j in range(start, end + sw, sw):
        n += 1
        data[i][j] = n
    # ===== for j
    start, end = end, start
    sw *= -1
# ===== for i

for i in range(row):
    for j in range(col):
        print('{0:3d} '.format(data[i][j]), end='')
    print()
```

```python
row, col = map(int, input('2차원 리스트를 만들려하는 행과 열의 개수: ').split())
data = [[0] * col for i in range(row)]

n = 0
start = [i for i in range(col)]
end = [i for i in range(col - 1, -1, -1)]

for i in range(row):
    for j in start:
        n += 1
        data[i][j] = n
    # ===== for j
    start, end = end, start
# ===== for i

for i in range(row):
    for j in range(col):
        print('{0:3d} '.format(data[i][j]), end='')
    print()
```
