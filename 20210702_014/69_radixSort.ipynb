```python
data = [19, 2, 31, 45, 30, 11, 121, 27]
isSort = False #    
radix = 1      #      

#     .
# isSort False        True     .
while not isSort:
    isSort = True
    #   ()     .
    queueList = [[] for i in range(10)]
    print('radix: {}'.format(radix))
    #      .
    for n in data:
        print('n: {0:3d}'.format(n), end=' => ')
        #   (radix  ) .
        digit = n // radix % 10
        print(digit)
        #   .
        queueList[digit].append(n)
        #    .
        if isSort and digit > 0:
            isSort = False
    # ===== for n in data =>     
    
    #    0    data  .
    index = 0
    # queueList  0  numbers      9  numbers    
    #  .
    for numbers in queueList:
        #         data  .
        for number in numbers:
            data[index] = number
            index += 1
        # ===== for number in numbers =>  1    data  .
    # ===== for numbers in queueList
    print(data)
    print('=' * 80)
    radix *= 10
```
