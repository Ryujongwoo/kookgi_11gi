데이터에 누락값이 없으면 좋겠지만 누락값은 대부분의 데이터에 존재하고 깔끔한 데이터를 위해 누락값을 처리하는 것인 매우 중요하다.  
누락값은 NaN으로 표기되며 누락값을 파이썬에서 사용하려면 넘파이 라이브러리가 필요하다.

```python
# 누락값을 사용하기 위해 넘파이 누락값을 불러온다.
from numpy import NaN, NAN, nan
```

```python
# 누락값은 0, ''과 같은 값과는 다른 개념으로 데이터 자체가 없다는 것을 의미한다. => NULL => '같다'라는 개념도 존재하지 않는다.
print(NaN == True)
print(NaN == False)
print(NaN == 0)
print(NaN == '')
```

```python
# 누락값은 값 자체가 없기 때문에 자기 자신과 비교해도 False가 출력된다.
print(NaN == NaN)
print(NAN == NAN)
print(nan == nan)
```

```python
# 판다스에는 누락값을 확인하는 isnull(), isna() 함수가 있고 누락값이면 True, 누락값이 아니면 False를 리턴한다.
print(pd.isnull(NaN))
print(pd.isna(nan))
print(pd.isnull(0))
print(pd.isna(0))
print(pd.isnull(''))
print(pd.isna(''))
```

```python
# Cases_* => 발병자 수, Deaths_* => 사망자 수
ebola = pd.read_csv('./data/country_timeseries.csv')
ebola
```

```python
# 넘파이의 count_nonzero() 함수는 배열에서 0이나 False가 아닌 데이터의 개수를 센다.
print(np.count_nonzero([False, 0]))
print(np.count_nonzero([False, 0, True, 1, 999, NaN]))
# isnull(), isna() 함수는 누락값이면 True, 누락값이 아니면 False를 리턴한다.
print(np.count_nonzero(ebola.isnull()))
print(np.count_nonzero(ebola.isna()))
print(np.count_nonzero(ebola['Cases_Guinea'].isnull()))
print(np.count_nonzero(ebola.Cases_Guinea.isna()))
# count() 함수는 NaN이 아닌 데이터의 개수를 센다.
print(ebola['Cases_Guinea'].count())
```

```python
# count() 함수로 NaN이 아닌 데이터의 개수를 알 수 있다.
print(ebola.count())
```

```python
print(ebola.shape)
print('ebola 데이터는 {}행 {}열 입니다.'.format(ebola.shape[0], ebola.shape[1]))
num_row = ebola.shape[0]
num_missing = num_row - ebola.count()
num_missing
```

```python
# value_counts() 함수는 지정한 열의 데이터 출현 빈도수를 얻어온다. => 기본값으로 NaN은 무시하고 출현 빈도수를 계산한다.
# print(ebola['Cases_Guinea'].value_counts())
print(ebola.Cases_Guinea.value_counts())
```

```python
# value_counts() 함수에 dropna 옵션을 False로 지정하면 NaN도 출현 빈도수를 계산한다.
print(ebola.Cases_Guinea.value_counts(dropna=False))
```

```python
# 기초 통계 계산 함수의 skipna 옵션의 기본값은 True 이고 NaN 값을 무시한다. NaN도 취급하게 하려면 skipna 옵션을 False로
# 지정하면 된다.
print(ebola.Cases_Guinea.sum())
print(ebola.Cases_Guinea.sum(skipna=True))
print(ebola.Cases_Guinea.sum(skipna=False))
print(ebola.Cases_Guinea.mean())
print(ebola.Cases_Guinea.max())
print(ebola.Cases_Guinea.min())
print(ebola.Cases_Guinea.count())
# print(ebola.Cases_Guinea.count(skipna=True)) # count() 함수는 skipna 옵션을 사용할 수 없다.
# print(ebola.Cases_Guinea.count(skipna=False))
```

누락값 처리하기

```python
# ebola.loc[:10, ['Date', 'Day', 'Cases_Guinea', 'Cases_Liberia', 'Cases_SierraLeone']]
ebola.iloc[:10, :5]
```

```python
# fillna() 함수의 인수로 특정 데이터를 지정하면 누락값을 지정된 값으로 변경한다.
ebola.fillna(0).iloc[:10, :5]
```

```python
# fillna() 함수의 method 옵션을 'ffill'로 지정하면 누락값이 나타나기 바로 전의 값으로 변경한다.
ebola.fillna(method="ffill").iloc[:10, :5]
```

```python
# fillna() 함수의 method 옵션을 'bfill'로 지정하면 누락값이 나타난 바로 다음 값으로 변경한다.
ebola.fillna(method="bfill").iloc[:10, :5]
```

```python
# interpolate() 함수는 누락값 앞, 뒤의 데이터를 이용해서 중간값(평균값)으로 누락값을 변경한다.
ebola.interpolate().iloc[:10, :5]
```

```python
# dropna() 함수로 누락값이 포함된 행을 삭제할 수 있다.
ebola.dropna()
```
