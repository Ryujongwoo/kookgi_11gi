인덱싱

```python
# 배열의 각 요소는 axis 인덱스로 참조할 수 있다.
# 1차원 배열은 1개의 인덱스, 2차원 배열은 2개의 인덱스, 3차원 배열은 3개의 인덱스 요소를 참조할 수 있다.
# 인덱싱 작업시 [행][열]과 같이 인덱스를 지정하거나 [행, 열]과 같이 인덱스를 지정할 수 있다.
```

```python
# 1차원 배열
a = np.arange(24)
print(a)
# 1차원 배열 요소 참조 및 변경
print(a[5])
a[0] = 10000
print(a)
```

```python
# 2차원 배열
a = np.arange(24).reshape(4, 6)
print(a)
# 2차원 배열 요소 참조 및 변경
print(a[0][1])
print(a[0, 1])
a[0][1] = 10000
a[2, 4] = 20000
print(a)
```

```python
# 3차원 배열
a = np.arange(24).reshape(2, 4, 3)
print(a)
# 3차원 배열 요소 참조 및 변경
print(a[0][2][1])
print(a[1, 1, 2])
a[0][2][1] = 99999
a[1, 1, 2] = 55555
print(a)
```

배열 변환  
배열을 변환하는 방법은 전치, 배열 shape 변환, 배열 요소 추가, 배열 결합, 배열 분리 등이 있다.

```python
# 전치는 행렬의 인덱스가 바뀌는 것이다.
Image('./numpyImages/numpyImage08.jpg', width=200)
```

```python
# 넘파이에서 배열을 전치하기 위해 배열.T 속성을 사용한다.
a = np.arange(1, 7).reshape(2, 3)
print(a)
print(a.T) # 소문자 't'를 사용하면 에러가 발행된다.
```

```python
# 배열 형태 변경
# 넘파이는 배열의 shape을 변경하는 reshape() 함수와 ravel() 함수를 제공한다.
# reshape() 함수는 데이터 변경없이 지정된 shape으로 배열의 형태를 변환한다.
# ravel() 함수는 배열의 shape을 무조건 1차원으로 변환한다.
```

```python
a = np.random.randint(1, 10, (3, 4))
print(a)
print(a.reshape(2, 6)) # 단순히 reshape() 함수만 실행하면 배열의 구조가 변경되지 않는다.
print(a)
a = a.reshape(2, 6)    # reshape() 함수로 배열의 구조를 변경하려면 실행 결과를 다시 저장시켜야 한다.
print(a)
print(a.ravel())       # 단순히 ravel() 함수만 실행하면 배열의 구조가 변경되지 않는다.
print(a)
a = a.ravel()          # ravel() 함수로 배열의 구조를 변경하려면 실행 결과를 다시 저장시켜야 한다.
print(a)
```

```python
# ravel() 함수도 인덱싱, 슬라이싱의 경우처럼 원본 배열의 view를 보여준다.
# ravel() 함수가 반환한 배열을 수정하면 원본 배열의 내용이 같이 수정된다.
a = np.random.randint(1, 10, (3, 4))
print(a)
b = a.ravel()
print(b)
b[0] = 99999
print(b)
print(a)
```

```python
# reshape() 함수도 인덱싱, 슬라이싱의 경우처럼 원본 배열의 view를 보여준다.
# reshape() 함수가 반환한 배열을 수정하면 원본 배열의 내용이 같이 수정된다.
a = np.random.randint(1, 10, (3, 4))
print(a)
b = a.reshape(3, 2, 2)
print(b)
b[1][0][1] = 99999
print(b)
print(a)
```

```python
# resize() 함수는 배열의 shape과 크기를 변경한다.
# np.resize(배열, new_shape), 배열.resize(new_shape)
# resize()와 reshape() 함수는 shape을 변경한다는 부분에서는 동일하지만 reshape() 함수는 요소의 개수가 변경되면 에러가 발생되는
# 반면에 resize() 함수는 shape을 변경하는 과정에서 요소의 개수를 늘이거나 줄일 수 있다.
```

```python
a = np.random.randint(1, 10, (2, 6))
print(a)
print(np.resize(a, (6, 2))) # resize() 함수는 변경된 데이터를 사용하려면 실행한 결과를 새 배열에 저장시켜 사용해야 한다.
print(a)
b = np.resize(a, (6, 2))
print(b)
```

```python
# 요소의 개수가 변경되는 resize()
a = np.random.randint(1, 10, (2, 6))
print(a)
# a.reshape(2, 10) # reshape() 함수는 원본 요소의 개수와 reshape() 함수로 변경하는 요소의 개수가 다르면 에러가 발생된다.
# np.resize(배열, new_shape) 형식은 실행 결과 원본 데이터가 변경되지 않기 때문에 다른 변수에 저장해서 사용한다.
# 배열 요소의 개수가 12개에서 20개로 늘어나고 늘어난 요소에는 원본 내용이 반복되서 채워진다.
b = np.resize(a, (2, 10))
print(b)
# 배열.resize(new_shape) 형식은 실행하면 원본 데이터가 변경되므로 원본 배열 그대로 사용하면 된다.
# 배열 요소의 개수가 12개에서 20개로 늘어나고 늘어난 요소에는 0이 채워진다.
a.resize(2, 10)
print(a)
```

```python
a = np.random.randint(1, 10, (2, 6))
print(a)
b = np.resize(a, (2, 4)) # 배열 오소의 개수가 12개에서 8개로 줄어들고 줄어든 요소의 데이터는 분실된다.
print(b)
a.resize(2, 4)           # 배열 오소의 개수가 12개에서 8개로 줄어들고 줄어든 요소의 데이터는 분실된다.
print(a)
```

```python
# append() 함수는 배열 끝에 데이터(배열)를 추가한다. axis로 데이터가 추가되는 방향을 지정할 수 있다.
# np.append(배열, 추가할 데이터, axis)
```

```python
a = np.arange(1, 10).reshape(3, 3)
print(a)
b = np.arange(10, 19).reshape(3, 3)
print(b)
```

```python
# axis를 지정하지 않을 경우
# axis를 지정하지 않으면 배열은 1차원 배열로 변환되서 결합된다.
# result = np.append(a, b, axis=None)
result = np.append(a, b)
print(result)
```

```python
# axis=0으로 지정한 경우
# 배열의 아래쪽에 배열이 추가(열방향)되기 때문에 행의 개수는 달라도 상관없지만 열의 개수는 반드시 같아야 한다.
result = np.append(a, b, axis=0)
print(result)

# print(result.shape)    # 2차원 배열에서 shape은 행과 열을 의미한다.
# print(result.shape[0]) # 2차원 배열에서 shape[0]은 행을 의미한다.
# print(result.shape[1]) # 2차원 배열에서 shape[1]은 열을 의미한다.

# axis=0으로 설정시 shape[0]은 서로 달라도 상관없지만 shape[1]은 반드시 같아야 오류가 발생되지 않는다.
c = np.arange(19, 31).reshape(3, 4)
print(c)
result = np.append(a, c, axis=0)
print(result)
```

```python
# axis=1로 지정한 경우
# 배열의 오른쪽에 배열이 추가(행방향)되기 때문에 열의 개수는 달라도 상관없지만 행의 개수는 반드시 같아야 한다.
result = np.append(a, b, axis=0)
print(result)

# axis=1로 설정시 shape[1]은 서로 달라도 상관없지만 shape[0]은 반드시 같아야 오류가 발생되지 않는다.
c = np.arange(19, 31).reshape(4, 3)
print(c)
result = np.append(a, c, axis=1)
print(result)
```

```python
# insert() 함수는 지정된 index 위치에 데이터를 삽입한다. axis로 데이터가 삽입되는 방향을 지정할 수 있다.
# np.insert(배열, index, 추가할 데이터, axis)
a = np.arange(1, 10).reshape(3, 3)
print(a)
```

```python
# axis를 지정하지 않으면 1차원 배열로 변환하고 index 위치에 데이터를 삽입한다.
# print(np.insert(a, 1, 99999, axis=None))
print(np.insert(a, 1, 77777))
```

```python
# axis=0으로 지정하면 열 방향 index 전체에 데이터를 삽입한다.
print(np.insert(a, 1, 99999, axis=0))
```

```python
# axis=1로 지정하면 행 방향 index 전체에 데이터를 삽입한다.
print(np.insert(a, 1, 99999, axis=1))
```

```python
# delete() 함수는 지정된 index의 데이터를 삭제한다. axis로 데이터가 삭제되는 방향을 지정할 수 있다.
# delete(배열, index, axis)
a = np.arange(1, 10).reshape(3, 3)
print(a)
```

```python
# axis를 지정하지 않으면 1차원 배열로 변환하고 index 번째 데이터를 삭제한다.
# print(np.delete(a, 1, axis=None))
print(np.delete(a, 1))
```

```python
# axis=0으로 지정하면 열 방향으로 index 위치의 데이터가 삭제된다.
print(np.delete(a, 1, axis=0))
```

```python
# axis=1으로 지정하면 행 방향으로 index 위치의 데이터가 삭제된다.
print(np.delete(a, 1, axis=1))
```

배열 분리  
넘파이는 수평 방향으로 분할하는 hsplit() 함수와 배열을 수직 방향으로 분할하는 vsplit() 함수를 제공한다.

```python
# hsplit() 함수로 배열을 수평 방향으로 분할한다.
# np.hsplit(배열, 분할할 배열의 개수), np.hsplit(배열, [분할할 인덱스])
a = np.arange(1, 25).reshape(4, 6)
print(a)
```

```python
result = np.hsplit(a, 2) # hsplit() 함수의 실행 결과는 넘파이 배열이 저장된 파이썬 리스트이다.
print(type(result))      # <class 'list'>
print(result)
print(type(result[0]))   # <class 'numpy.ndarray'>
print(result[0])
Image('./numpyImages/numpyImage09.jpg', width=800)
```

```python
result = np.hsplit(a, 3)
print(type(result))
print(result)
print(type(result[0]))
print(result[0])
Image('./numpyImages/numpyImage10.jpg', width=800)
```

```python
result = np.hsplit(a, [1, 3, 5])
print(type(result))
print(result)
print(type(result[0]))
print(result[0])
Image('./numpyImages/numpyImage11.jpg', width=800)
```

```python
# vsplit() 함수로 배열을 수직 방향으로 분할한다.
# np.vsplit(배열, 분할할 배열의 개수), np.vsplit(배열, [분할할 인덱스])
a = np.arange(1, 25).reshape(4, 6)
print(a)
```

```python
result = np.vsplit(a, 2)
print(type(result))
print(result)
print(type(result[0]))
print(result[0])
Image('./numpyImages/numpyImage12.jpg', width=800)
```
