```python
a = [[0] * 5 for i in range(5)]
n = 0  # 1 ~ 25까지 1씩 증가하는 변수
s = 1  # 행과 열이 1또는 -1씩 증가 및 감소 처리를 하는 변수
i = 0  # 행
j = -1 # 열
k = 5  # 반복문의 반복횟수 제어에 사용하는 변수

while True:
    # 행 방향으로 숫자를 채운다.
    for p in range(1, k + 1):
        n += 1
        j += s
        a[i][j] = n
    # ===== for p
    k -= 1
    if k == 0:
        break
    # 열 방향으로 숫자를 채운다.
    for p in range(1, k + 1):
        n += 1
        i += s
        a[i][j] = n
    # ===== for p
    s *= -1
# ===== while True

for i in range(len(a)):
    for j in range(len(a)):
        print('{0:3d} '.format(a[i][j]), end='')
    print()
```

```python
a = [[0] * 5 for i in range(5)]
n = 1; a[2][2] = n; s = 1; i = 2; j = 2; k = 1; flag = False

while True:
    k += 1 # 3시 방향부터 숫자를 채운다.
    for p in range(1, k):
        n += 1; j += s
        if n > 25:
            flag = True
            break
        a[i][j] = n
    # ===== for p
    if flag:
        break
    for p in range(1, k):
        n += 1; i += s
        a[i][j] = n
    # k += 1 # 9시 방향부터 숫자를 채운다.
    s *= -1
# ===== while True

for i in range(len(a)):
    for j in range(len(a)):
        print('{0:3d} '.format(a[i][j]), end='')
    print()
```
