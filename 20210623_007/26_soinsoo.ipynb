```python
#            .
while True:
    n = int(input('  : '))
    # 2    .
    if n < 2:
        # print('2   .')
        break
    # ===== if n < 2
    # 2     .
    s = [] #      .
    c = 0  #  
    number = n
    
    #   .
    #             .
    while True:
        k = 2 #     .
        while True:
            r = n % k
            if r == 0: #   ?
                break
            # ===== if r == 0
            k += 1
        # ===== while True
        c += 1      #      1. =>   1 .
        s.append(k) #   .
        #        () n .
        n //= k
        if n == 1:  #   ?
            break
    # ===== while True
    
    #       .
    if c == 1:
        print(': {} => '.format(number))
    else:
        print(': {} => '.format(number), end='')
        for i in range(c - 1):
            print(s[i], end=' * ')
        print(s[-1])
# ===== while True
print(': {} =>  '.format(n))
```
