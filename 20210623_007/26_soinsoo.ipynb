```python
# 소인수 분해가 불가능한 숫자가 입력될 때 까지 실행해야 하기 때문에 무한루프를 돌린다.
while True:
    n = int(input('소인수 분해할 수: '))
    # 2미만인 숫자가 입력되면 프로그램을 종료한다.
    if n < 2:
        # print('2이상인 숫자를 입력해야 합니다.')
        break
    # ===== if n < 2
    # 2이상인 숫자가 입력되면 소인수 분해를 실행한다.
    s = [] # 분해된 소인수를 기억할 빈 리스트를 만든다.
    c = 0  # 소인수의 개수
    number = n
    
    # 소인수 분해를 시작한다.
    # 소인수 분해를 하기위해서 입력되는 숫자의 크기에 따라 반복 횟수가 달라지므로 무한루프를 돌려 처리한다.
    while True:
        k = 2 # 소인수 분해를 시작할 수를 초기화한다.
        while True:
            r = n % k
            if r == 0: # 소인수 분해가 되었는가?
                break
            # ===== if r == 0
            k += 1
        # ===== while True
        c += 1      # 소인수 분해가 되었으므로 소인수의 개수를 1증가시킨다. => 소인수의 개수가 1개면 소수이다.
        s.append(k) # 소인수를 리스트에 추가한다.
        # 다음 소인수를 계산하기 위해서 소인수 분해되고 남은 숫자(몫)을 n에 저장시킨다.
        n //= k
        if n == 1:  # 소인수 분해가 끝났는가?
            break
    # ===== while True
    
    # 소인수 분해를 하기위해 입력받은 숫자가 소수인가 비교한다.
    if c == 1:
        print('입력: {} => 소수'.format(number))
    else:
        print('입력: {} => '.format(number), end='')
        for i in range(c - 1):
            print(s[i], end=' * ')
        print(s[-1])
# ===== while True
print('입력: {} => 프로그램 종료'.format(n))
```
