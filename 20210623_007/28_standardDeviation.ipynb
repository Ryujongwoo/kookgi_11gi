n개의 데이터(999가 입력되기 전 까지)를 입력받아 최대값과 최소값을 제외한 나머지 데이터의 표준편차를 계산하는 프로그램
***
$$\sqrt{\frac{\sum_{i=1}^{N-2} (x_i - \bar{x})^2}{n-2}}$$

```python
# 최대값, 최소값 판단하기 - 1
data = [35, 91, 75, 62, 84]
# 최대값을 기억할 변수는 아주 작은 값을 초기치로 지정하고 최소값을 기억할 변수는 아주 큰 값을 초기치로 넣어주고 비교한다.
b = 0
s = 100
for score in data:
    if score > b:
        b = score
    if score < s:
        s = score
print('최대값: {}, 최소값: {}'.format(b, s))
```

```python
# 최대값, 최소값 판단하기 - 2
data = [35, 91, 75, 62, 84]
# 최대값을 기억할 변수와 최소값을 기억할 변수에 리스트의 0번째 인덱스의 값을 초기치로 넣어주고 비교한다.
b = data[0]
s = data[0]
for score in data:
    if score > b:
        b = score
    elif score < s:
        s = score
print('최대값: {}, 최소값: {}'.format(b, s))
```

```python
# 최대값, 최소값 판단하기 - 3
data = [35, 91, 75, 62, 84]
data.sort()
print('최대값: {}, 최소값: {}'.format(data[-1], data[0]))
data.sort(reverse=True)
print('최대값: {}, 최소값: {}'.format(data[0], data[-1]))
```

```python
# 최대값, 최소값 판단하기 - 4
data = [35, 91, 75, 62, 84]
print('최대값: {}, 최소값: {}'.format(max(data), min(data)))
```

```python
# 표준편차를 계산할 데이터를 999가 입력될 때 까지 입력받는다.
score = [] # 표준편차를 계산할 데이터를 기억할 빈 리스트를 선언한다.
# 999가 몇 번째 데이터로 입력될지 모르기 때문에 무한루프를 돌린다.
while True:
    data = int(input('{}번째 표준편차를 계산할 데이터를 입력하세요: '.format(len(score) + 1)))
    if data == 999:    # 입력받은 데이터가 999면 더 이상 데이터를 입력받을 필요가 없기때문에 무한루프를 탈출한다.
        break
    score.append(data) # 입력받은 데이터를 리스트에 추가한다.
print('입력받은 데이터의 개수는 {}개 이고 데이터는 {} 입니다.'.format(len(score), score))
```

```python
big = max(score)
small = min(score)
score.sort()
```

최대값과 최소값을 제외한 나머지 데이터의 평균을 계산한다.

```python
total = 0
for s in score:
    total += s
# total = total - big - small
total -= big + small
print('최대값과 최소값을 제외한 나머지 데이터의 합계: {}'.format(total))
mean = total / (len(score) - 2)
print('최대값과 최소값을 제외한 나머지 데이터의 평균: {}'.format(mean))
```

```python
total = 0
for i in range(1, len(score) - 1):
    total += score[i]
print('최대값과 최소값을 제외한 나머지 데이터의 합계: {}'.format(total))
mean = total / (len(score) - 2)
print('최대값과 최소값을 제외한 나머지 데이터의 평균: {}'.format(mean))
```

```python
total = 0
for s in score[1:len(score) - 1]:
    total += s
print('최대값과 최소값을 제외한 나머지 데이터의 합계: {}'.format(total))
mean = total / (len(score) - 2)
print('최대값과 최소값을 제외한 나머지 데이터의 평균: {}'.format(mean))
```

최대값과 최소값을 제외한 나머지 데이터와 산술평균의 편차의 제곱의 합계를 계산한다.

```python
total = 0
for s in score[1:len(score) - 1]:
    total += (s - mean) ** 2
print('최대값과 최소값을 제외한 나머지 데이터와 산술평균의 편차의 제곱의 합계: {}'.format(total))
```

분산(최대값과 최소값을 제외한 나머지 데이터와 산술평균의 편차의 제곱의 합계의 산술평규)을 계산한다.

```python
var = total / (len(score) - 2)
print('최대값과 최소값을 제외한 데이터의 분산: {}'.format(var))
```

표준편차(분산의 제곱근)를 계산한다.

```python
std = var ** 0.5
print('최대값과 최소값을 제외한 데이터의 표준편차: {}'.format(std))
```

```python
import math
print('최대값과 최소값을 제외한 데이터의 표준편차: {}'.format(math.sqrt(var)))
```
